**Scenario:**
You've surely seen a seven-segment display. It's a device (sometimes electronic, sometimes mechanical) designed to present one decimal digit using a subset of seven segments. If you're not familiar with it, refer to the following [Wikipedia article](https://en.wikipedia.org/wiki/Seven-segment_display).

Your task is to write a program that simulates the work of a seven-segment display, but using single LEDs instead of segments.

Each digit is constructed from 13 LEDs (some lit, some dark, of course). Hereâ€™s how we imagine it:

```
  # ### ### # # ### ### ### ### ### ### 
  #   #   # # # #   #     # # # # # # # 
  # ### ### ### ### ###   # ### ### # # 
  # #     #   #   # # #   # # #   # # # 
  # ### ###   # ### ###   # ### ### ###
```

Note: The number 8 shows all the LED lights on.

Your code should display any non-negative integer number entered by the user.

**Tip:** Using a list containing patterns of all ten digits may be very helpful.
